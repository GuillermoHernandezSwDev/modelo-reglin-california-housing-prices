```python
#SEPARAR LAS CARACTERSTICAS DE LA ETIQUETA

X = final_df.drop(["median_house_value"], axis=1)
y = final_df["median_house_value"]
```

```python
#SEPARAR LOS DATOS EN 2 PARTES:
    #-Conjunto de entrenamiento
    #-Conjunto de prueba

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
comp = {"Prediccion" : predictions, "Precio" : y_test}
pd.DataFrame(comp)
```

```python
#EVALUACIN - OVERFITTING (Sobreajuste)

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
#Error

from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)
```
