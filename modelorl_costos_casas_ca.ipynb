```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sb
```

```python
data_frame = pd.read_csv("cal_housing_pri.csv")
```

```python
data_frame.info()
```

```python
data_frame.head(10)
```

```python
data_frame2 = data_frame.dropna()
```

```python
data_frame2.info()
```

```python
data_frame2["ocean_proximity"].value_counts
```

```python
df_wdummies = pd.get_dummies(data_frame2["ocean_proximity"], dtype=int)
```

```python
df_wdummies.head(10)
```

```python
data_frame3 = data_frame2.join(df_wdummies)
```

```python
data_frame3.head(10)
```

```python
final_df = data_frame3.drop(["ocean_proximity"], axis=1)
```

```python
final_df.head(10)
```

```python
sb.set_theme(rc={'figure.figsize': (15,10)})
sb.heatmap(final_df.corr(), annot=True, cmap="YlGnBu")
```

```python
final_df.corr()["median_house_value"].sort_values(ascending=False)
```
